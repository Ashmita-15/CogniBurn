```python
df['travelling_time'].unique()
```

```python
df['travelling_time'] = df['travelling_time'].str.lower().map({
    '30 - 60 minutes': 0.45,
    '0 - 30 minutes': 0.15,
    '1 - 1.30 hour': 1.15,
    '2 - 2.30 hour':2.15,
    '1.30 - 2 hour':1.45,
    '2.30 - 3 hour':2.5,
    'more than 3 hour': 3.5
})
```

# Dataset 2: Stress Factors

```python
df1=pd.read_csv('datasets/Stress_Factor.csv')
df1.sample(10)
```

```python
df1.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df1.rename(columns={
    'Kindly Rate your Sleep Quality 😴': 'sleep_quality',
    'How many times a week do you suffer headaches 🤕?': 'headache_freq',
    'How would you rate you academic performance 👩‍🎓?': 'academic_performance',
    'how would you rate your study load?': 'study_load',
    'How many times a week you practice extracurricular activities 🎾?': 'extracurricular_freq',
    'How would you rate your stress levels?': 'stress_level'
}, inplace=True)
df1.sample(10)
```

```python
df1['sleep_quality'].unique()
```

```python
scaler = MinMaxScaler()
cols = [
    'sleep_quality',
    'headache_freq',
    'academic_performance',
    'study_load',
    'extracurricular_freq',
    'stress_level'
]

df1[cols] = scaler.fit_transform(df1[cols])
```

```python
df1.sample(10)
```

```python
df1['mental_strain'] = (df1['stress_level'] + df1['study_load'] + df1['headache_freq'])
df1['recovery_score']=df1['sleep_quality']+df1['extracurricular_freq']
df1['performance_pressure']=(df1['study_load']*(1-df1['academic_performance']))
```

```python
df1['burnout_score'] = (0.4 * df1['mental_strain'] +0.3 * (1 - df1['recovery_score']) +0.3 * df1['performance_pressure'])
df1['burnout_level'] = pd.qcut(
    df1['burnout_score'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

# Final Dataset:

### Combining all three datasets

```python
n=1000
data = pd.DataFrame()
```

```python
data['study_time'] = df['study_time'].sample(n, replace=True).values
data['social_media_usage'] = df['social_media_usage'].sample(n, replace=True).values
```

```python
data['sleep_quality'] = df1['sleep_quality'].sample(n, replace=True).values
data['headache_freq'] = df1['headache_freq'].sample(n, replace=True).values
data['study_load'] = df1['study_load'].sample(n, replace=True).values
data['extracurricular_freq'] = df1['extracurricular_freq'].sample(n, replace=True).values
```

```python
data['screen_time'] = df2['screen_time_(hrs/day)'].sample(n, replace=True).values
data['sleep_hours'] = df2['sleep_duration_(hrs)'].sample(n, replace=True).values
data['physical_activity'] = df2['physical_activity_(hrs/week)'].sample(n, replace=True).values
data['exam_anxiety'] = df2['anxious_before_exams'].sample(n, replace=True).values
```

```python
combined_stress = pd.concat([
    df['stress_level'],
    df1['stress_level'],
    df2['stress_level']
])
data['stress_level'] = combined_stress.sample(n, replace=True).values
data['stress_level'] = (
    data['stress_level'] + (1 - data['sleep_hours'])
) / 2
```

```python
data.sample(20)
```

```python
scaler = MinMaxScaler()

cols = [
    'study_time', 'social_media_usage',
    'sleep_quality', 'headache_freq',
    'study_load', 'extracurricular_freq',
    'screen_time', 'sleep_hours',
    'physical_activity', 'stress_level'
]

data[cols] = scaler.fit_transform(data[cols])
```

```python
#Mental strain
data['mental_strain'] = (
    data['stress_level'] +
    data['study_load'] +
    data['headache_freq']
)

# Recovery score
data['recovery_score'] = (
    data['sleep_quality'] +
    data['extracurricular_freq']
)

# Performance pressure
data['performance_pressure'] = (
    data['study_load'] * (1 - data['study_time'])
)

# Lifestyle stress
data['lifestyle_stress'] = (
    data['screen_time'] + (1 - data['sleep_hours'])
)
```

```python
data['burnout_score'] = (
    0.4 * data['mental_strain'] +
    0.3 * (1 - data['recovery_score']) +
    0.3 * data['performance_pressure']
)
```

```python
data['burnout_level'] = pd.qcut(
    data['burnout_score'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```
