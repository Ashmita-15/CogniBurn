```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

# Dataset 1: Student Behaviour

```python
df= pd.read_csv('datasets/Student_Behaviour.csv')
df.sample(10)
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "_")
df.columns = df.columns.str.lower() 
df.rename(columns={
    'social_medai_&_video': 'social_media_usage',
    'daily_studing_time': 'study_time',
    'college_mark': 'college_marks'
}, inplace=True)
df.sample(10)
```

```python
df.isnull().sum()
```

```python
binary_cols=[
    'certification_course',
    'do_you_like_your_degree?',
    'part-time_job'
]
for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df['financial_status'].unique()
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
df['financial_status'] = df['financial_status'].map({
    'Fabulous':2,
    'good':4,
    'Bad':7,
    'Awful':9
})
```

```python
df['study_time'].unique()
```

```python
df['study_time']=df['study_time'].map({
    '0 - 30 minute':0.25,
    '30 - 60 minute':0.75,
    '1 - 2 Hour':1.5,
    '2 - 3 hour':2.5,
    '3 - 4 hour':3.5,
    'More Than 4 hour':5
})
```

```python
df['stress_level'].unique()
```

```python
df['stress_level']=df['stress_level'].map({
    'fabulous':2,
    'Good':4,
    'Bad':7,
    'Awful':9
})
```

```python
df['social_media_usage'].unique()
```

```python
df['social_media_usage'] = df['social_media_usage'].map({
    '1.30 - 2 hour': 1.75,
    '1 - 1.30 hour': 1.25,
    'More than 2 hour': 3,
    '30 - 60 Minute': 0.75,
    '1 - 30 Minute': 0.5,
    '0 Minute': 0
})
```

```python
df['travelling_time'].unique()
```

```python
df['travelling_time'] = df['travelling_time'].str.lower().map({
    '30 - 60 minutes': 0.45,
    '0 - 30 minutes': 0.15,
    '1 - 1.30 hour': 1.15,
    '2 - 2.30 hour':2.15,
    '1.30 - 2 hour':1.45,
    '2.30 - 3 hour':2.5,
    'more than 3 hour': 3.5
})
```

```python
df['willingness_to_pursue_a_career_based_on_their_degree'].unique()
```

```python
df['career_alignment'] = df['willingness_to_pursue_a_career_based_on_their_degree'].str.replace('%', '').astype(int)
df['career_alignment'] = df['career_alignment'] / 100
```

```python
df.drop([
    'height(cm)',
    'weight(kg)',
    'hobbies',
    'salary_expectation',
    'prefer_to_study_in',
    'department',
    'willingness_to_pursue_a_career_based_on_their_degree'
], axis=1, inplace=True)
```

```python
# New column: Academic pressure
df['academic_pressure'] = (
    df['study_time'] * (100 - df['college_marks'])
)

# New column: Lifestyle stress
df['lifestyle_stress'] = (
    df['social_media_usage'] + df['travelling_time']
)
```

```python
df.drop([
    '10th_mark',
    '12th_mark',
    'certification_course'
], axis=1, inplace=True)
```

```python
df.sample(10)
```

# Dataset 2: Stress Factors

```python
df1=pd.read_csv('datasets/Stress_Factor.csv')
df1.sample(10)
```

```python
df1.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df1.rename(columns={
    'Kindly Rate your Sleep Quality 😴': 'sleep_quality',
    'How many times a week do you suffer headaches 🤕?': 'headache_freq',
    'How would you rate you academic performance 👩‍🎓?': 'academic_performance',
    'how would you rate your study load?': 'study_load',
    'How many times a week you practice extracurricular activities 🎾?': 'extracurricular_freq',
    'How would you rate your stress levels?': 'stress_level'
}, inplace=True)
df1.sample(10)
```

```python
df1['sleep_quality'].unique()
```

```python
scaler = MinMaxScaler()
cols = [
    'sleep_quality',
    'headache_freq',
    'academic_performance',
    'study_load',
    'extracurricular_freq',
    'stress_level'
]

df1[cols] = scaler.fit_transform(df1[cols])
```

```python
df1.sample(10)
```

```python
df1['mental_strain'] = (df1['stress_level'] + df1['study_load'] + df1['headache_freq'])
df1['recovery_score']=df1['sleep_quality']+df1['extracurricular_freq']
df1['performance_pressure']=(df1['study_load']*(1-df1['academic_performance']))
```

```python
df1['burnout_score'] = (0.4 * df1['mental_strain'] +0.3 * (1 - df1['recovery_score']) +0.3 * df1['performance_pressure'])
df1['burnout_level'] = pd.qcut(
    df1['burnout_score'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```
